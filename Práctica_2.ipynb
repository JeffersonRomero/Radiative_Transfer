```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_solar30=pd.read_csv('ssolar_goa30.txt',sep='\s+')
df_solar30['Direct_2']=np.cos(np.radians(30)) * df_solar30['Direct']
df_solar30
```

```python
df_solar60=pd.read_csv('solargoa_60.txt',sep='\s+')
df_solar60['Direct_2']=np.cos(np.radians(60)) * df_solar60['Direct']
df_solar60
```

```python
df_solar80=pd.read_csv('ssolar_goa80.txt',sep='\s+')
df_solar80['Direct_2']=np.cos(np.radians(80)) * df_solar80['Direct']
df_solar80
```

```python
df_lib30=pd.read_csv('output30.txt',sep='\s+',header=None)
df_lib30[1]=df_lib30[1]/1000
df_lib30[2]=df_lib30[2]/1000
df_lib30[3]=df_lib30[3]/1000
df_lib30[4]=df_lib30[4]/1000
df_lib30[5]=df_lib30[5]/1000
df_lib30
```

```python
df_lib60=pd.read_csv('output60.txt',sep='\s+',header=None)
df_lib60[1]=df_lib60[1]/1000
df_lib60[2]=df_lib60[2]/1000
df_lib60[3]=df_lib60[3]/1000
df_lib60[4]=df_lib60[4]/1000
df_lib60[5]=df_lib60[5]/1000
df_lib60
```

```python
df_lib80=pd.read_csv('output80.txt',sep='\s+',header=None)
df_lib80[1]=df_lib80[1]/1000
df_lib80[2]=df_lib80[2]/1000
df_lib80[3]=df_lib80[3]/1000
df_lib80[4]=df_lib80[4]/1000
df_lib80[5]=df_lib80[5]/1000
df_lib80
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_solar30['Wavelength'],df_solar30['Global'], label='Solar 30')
plt.plot(df_solar60['Wavelength'],df_solar60['Global'], label='Solar 60')
plt.plot(df_solar80['Wavelength'],df_solar80['Global'], label='Solar 80')
plt.plot(df_lib30[0],df_lib30[1], label='Lib 30')
plt.plot(df_lib60[0],df_lib60[1], label='Lib 60')
plt.plot(df_lib80[0],df_lib80[1], label='Lib 80')
plt.title('SSolarGOA vs Libradtran Global')
plt.xlabel(r'$Wavelenght (nm)$')
plt.ylabel(r'$W/m^2$')
plt.legend()
plt.show()
```
